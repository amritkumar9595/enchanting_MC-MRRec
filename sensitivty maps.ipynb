```python
items=os.listdir(root)
l2=os.listdir(root2)
print(l2)
items
```

```python
for item in tqdm_notebook(items):
    print("item=",item)
    if item not in l2:
        file = root+item
        try:
            print("file = ",file)
            hf = h5py.File(file)
        except:
            continue
        volume_kspace1 = hf['kspace'][()]
        volume_kspace =volume_kspace1[:,:,:,::2]+1j*volume_kspace1[:,:,:,1::2]
        volume_kspace=volume_kspace.transpose(0,3,1,2)
        sr = 0.85
        Nz = volume_kspace.shape[3]
        Nz_sampled = int(np.ceil(Nz*sr))
        volume_kspace[:,:,:,Nz_sampled:] = 0
    #     volume_kspace = volume_kspace[100:,:,:,Nz_sampled:]
        print('volume_kspace.shape',volume_kspace.shape)

        s_e_d = volume_kspace.shape[3]
        if s_e_d==170:
            sp_r5 = np.load("/media/student1/NewVolume/MR_Reconstruction/midl/MC-MRRec-challenge/Data/poisson_sampling/R5_218x170.npy")
        elif s_e_d==174:
            sp_r5 = np.load("/media/student1/NewVolume/MR_Reconstruction/midl/MC-MRRec-challenge/Data/poisson_sampling/R5_218x174.npy")
        elif s_e_d==180:
            sp_r5 = np.load("/media/student1/NewVolume/MR_Reconstruction/midl/MC-MRRec-challenge/Data/poisson_sampling/R5_218x180.npy")

    #     volume_kspace=volume_k
    #     volume_kspace.shape is(256,12,218,170)
        ##loop through slices
    #     plt.figure()
    #     for slice_kspace in volume_kspace[150:][:1]:
        volume_calib=[]
        i = 0
        for slice_kspace in volume_kspace:
            print("i = ",i)
            i = i+1
#             print("slice_ksp.shape",slice_kspace.shape)
            k_u = slice_kspace.copy()
#             k_u[:,~sp_r5[0]] = 0
            img_und=np.fft.ifft2(k_u)
            #shape of slice_kspace=(12,218,170)
            slice_image = np.fft.ifft2(slice_kspace,axes=(1,2))

    #         for ii in range(12):
    #             plt.imshow(np.abs(slice_image[ii,:,:]),cmap = "gray")
    #             plt.show()
    #             break

        # slice_kspace = volume_kspace[100] # Choosing the 20-th slice of this volume
            kspace1=np.expand_dims(np.transpose(k_u,axes=[1,2,0]),axis=0)
    #         calib = bart.bart(1, f'ecalib -d0 -m1', (kspace1)) ##
#             print('kpace1',kspace1.shape)
            [img,calib] = bart(2, f'nlinv -g',kspace1)
    #         print('calib',calib.shape)
    #         calib.shape==(1,218,170,12)
    #         plt.figure()
            slice_calib=[]
            for ii in range(calib.shape[3]):
    #             plt.figure()
    #             calib_slice=np.fft.ifftshift(np.abs(calib[0,:,:,ii]))
                calib_slice=np.fft.ifftshift(calib[0,:,:,ii])##
    #             print(calib_slice.shape)
                slice_calib.append(calib_slice)
    #             plt.imshow(calib_slice.astype(np.float32), cmap='gray')
    #             plt.show() 
    #             plt.imshow(np.abs(slice_image[ii,:,:]),cmap = "gray")
    #             plt.show() 

            volume_calib.append(np.array(slice_calib))

    #         break 
        volume_calib = np.array(volume_calib)  
    #     print(volume_calib.shape)
    #     print(item)
        with open(root2+item,'wb') as f:
            np.save(f,volume_calib)
    #     break
    else:
        print(item)

```
